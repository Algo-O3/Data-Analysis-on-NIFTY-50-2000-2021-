```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import seaborn as sns
import os
```

```python
dfs = []
file_names = []
path_name= 'data_set/*.csv'
csv_files = glob.glob(path_name)

for temp_csv_file in csv_files:
    df = pd.read_csv(temp_csv_file)
    dfs.append(df)
    file_names.append(os.path.basename(temp_csv_file))

```

```python
#Preprocessing the data
```

```python
for i in range(len(dfs)):
        dfs[i]['Date'] = pd.to_datetime(dfs[i]['Date'])
        dfs[i] = dfs[i].sort_values('Date')
        dfs[i] = dfs[i].drop(['Symbol', 'Series', 'Trades', 'Deliverable Volume', '%Deliverble'], axis=1, errors='ignore')
  
```

```python
#Generalising it for all stocks
```

```python
def predict_stock_prices(data_set, MSE, R2, Y_PREDICT,Y_TEST):
    for i, temp in enumerate(data_set):
        if temp.empty:
            print(f"DataFrame {i} is empty, skipping.")
            continue

        x_temp = temp.drop(['Date', 'Close'], axis=1)
        y_temp = temp['Close']

        if x_temp.empty or y_temp.empty:
            print(f"After dropping columns, DataFrame {i} resulted in empty data, skipping.")
            continue
        x_train, x_test, y_train, y_test = train_test_split(x_temp, y_temp, test_size=0.2, shuffle=False)

        if x_train.empty or y_train.empty:
            print(f"Train set is empty for DataFrame {i}, skipping.")
            continue
        model = LinearRegression()
        model.fit(x_train, y_train)
        y_predict = model.predict(x_test)
        mse = mean_squared_error(y_test, y_predict)
        r2 = r2_score(y_test, y_predict)
        Y_PREDICT.append(y_predict)
        MSE.append(mse)
        R2.append(r2)
        Y_TEST.append(y_test)
```

```python
mse_all=[]
r2_all=[]
y_predict_all=[]
y_test_all=[]
predict_stock_prices(dfs,mse_all,r2_all,y_predict_all,y_test_all)
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
dropdown = widgets.Dropdown(
    options=file_names,
    value=file_names[0],
    description='Select File:',
)
```

```python
output = widgets.Output()
```

```python
def update_output(change):
    with output:
        clear_output()
        selected_file = change['new']
        selected_index = file_names.index(selected_file)
        if mse_all and r2_all:
            print(f"MSE: {mse_all[selected_index]}")
            print(f"R2: {r2_all[selected_index]}")
            plt.figure(figsize=(10, 5))
            plt.plot(dfs[selected_index]['Date'], dfs[selected_index]['Close'], label='Actual Prices')
            plt.plot(dfs[selected_index]['Date'].iloc[-len(y_test_all[selected_index]):], y_predict_all[selected_index], label='Predicted Prices', linestyle='dashed')
            plt.legend()
            plt.title('Stock Price Prediction')
            plt.xlabel('Date')
            plt.ylabel('Price')
            plt.show()
        else:
            print("No valid data for prediction.")
```

```python
dropdown.observe(update_output, names='value')
```

```python
display(dropdown, output)
update_output({'new': dropdown.value})
```
